a)
----
different setups for Data Integration:
---
1) data warehousing: Bring the data together into a single repository:

pros :

- Data retrieval is faster within data warehouses.
- Query performance predictable and under your control.

cons:
 
-Preparation is very frequently time consuming for effort is needed to create a cohesive, compatible system of data collection, storage, and retrieval.
-Compatibility with existing systems. The use of data warehousing technology may require a company to modify the database system already in place.

2) Keep the data where it is, and send queries back and forth:

pros :

better fit for dynamic data Or when data warehousing is not feasible.

cons:

- Need two-way mappings, a query on the mediated schema
needs to be translated into queries over data source schemas
- Not as efficient and clean as data warehousing.

**in our case best to choose data warehousing.**

B)
------
key challenges for data integration:
----
1- Data extraction, reconciliation, and cleaning
- Get the data from each source in a structured form
- Often need to use wrappers to extract data from web sources
- May need to define a schema

2- Schema alignment and mapping
- Decide on the best mediated schema
- Figure out mappings and matchings between the local schemas and the global schema

3- Answer queries over the global schema
- Many sources have limits on how you can access them
- Limits on the number of queries you can issues (say 100 per min)
- Limits on the types of queries (e.g., must enter a zipcode to get information from a web source)
- In the second scenario, need to figure out how to map a query on global schema onto queries over local schemas
- Also need to decide which sources contain relevant data

4- Limitations in mechanisms for accessing sources
- Many sources have limits on how you can access them
- Limits on the number of queries you can issues (say 100 per min)
- Limits on the types of queries (e.g., must enter a zipcode to get information from a web source)

c)
----
SINGLE-SOURCE PROBLEMS
----
Depends largely on the source,
Databases can enforce constraints, whereas data extracted from files or
spreadsheets, or scraped from webpages is much more messy.
Types of problems:
----
- Ill-formatted data, especially from webpages or files or spreadsheets
- Missing or illegal values, Misspellings, Use of wrong fields, Extraction issues (not easy to separate out different fields)
- Duplicated records, Contradicting Information, Referential Integrity Violations
- Unclear default values (e.g., data entry software needs something)
- Evolving schemas or classification schemes (for categorical attributes)
- Outliers

**we can deal with all this issues with data cleaning.**

PART THREE: Hypothesis Testing 
---

my null hypo is H0= Mean(Male)- Mean(Female)=0
my alter hypo is H1= Mean(Male)- Mean(Female)!=0
I will set the varince by : unknown but assum to be equal,
and this two new sub-datasets are independant .
I will use T test because of the type of the variance and the sub-datasets

```python
#t-test formula
def Ttest_df(dfX,dfY):
    meanX=dfX['Salary'].mean()
    meanY=dfY['Salary'].mean()
    stdX=dfX['Salary'].std()
    stdY=dfY['Salary'].std()

    sizeOfx=len(dfX)
    sizeOfy=len(dfY)
    
    stand=(((sizeOfx-1)*(stdX**2))+((sizeOfy-1)*(stdY**2)))/(sizeOfx+sizeOfy-2)

    dOFf=(sizeOfx+sizeOfy-2)

    error=math.sqrt((stand/sizeOfx)+(stand/sizeOfy))

    t_test=((meanX-meanY)-0)/error
    return t_test
```

```python
#point a
dfMales = df[df['Gender'] == 'Male']
dfMales.head()
```

```python
#point a
dfFemales = df[df['Gender'] == 'Female']
dfFemales.head()
```

```python
#test using my function and online calculator 
#full dataframe
Ttest_df(dfMales,dfFemales)
#The p-value is .20916.
#The result is not significant at p < .05.
```

```python
#test using scipy to check
t_val,p_val=ttest_ind(dfMales.Salary,dfFemales.Salary);t_val,p_val #full
```

```python
#point b
dfMales=dfMales.sort_values('Age')
dfFemales=dfFemales.sort_values('Age')
```

```python
#point b
dfMalesUnder = dfMales[dfMales['Age'] < 35]

dfFemalesUnder = dfFemales[dfFemales['Age'] < 35]
dfMalesUnder.head()
```

```python
#test using my function and online calculator 
#under dataframe
Ttest_df(dfMalesUnder,dfFemalesUnder)
#The p-value is .816559.
#The result is not significant at p < .05.
```

```python
#test using scipy to check
t_valU,p_valU=ttest_ind(dfMalesUnder.Salary,dfFemalesUnder.Salary);t_valU,p_valU #under
```

```python
#point b

dfFemalesOver = dfFemales[dfFemales['Age'] >= 35]

dfMalesOver = dfMales[dfMales['Age'] >= 35]
dfMalesOver.head()
```

```python
#test using my function and online calculator 
#over dataframe
Ttest_df(dfMalesOver,dfFemalesOver)
#The p-value is .15937.
#The result is not significant at p < .05.
```

```python
t_valO,p_valO=ttest_ind(dfMalesOver.Salary,dfFemalesOver.Salary);t_valO,p_valO #over
```

comment:
---
in full, Under and over cases we can not reject the hypothesis because the T score is not 
bigger than 2 and all p test results are not significant at p < .05.
the only case that might have discrimination is the over 35 case.
