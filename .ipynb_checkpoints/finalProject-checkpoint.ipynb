PART THREE: Hypothesis Testing 
---

```python

def Ttest_df(dfX,dfY):
    meanX=dfX['Salary'].mean()
    meanY=dfY['Salary'].mean()
    stdX=dfX['Salary'].std()
    stdY=dfY['Salary'].std()

    sizeOfx=len(dfX)
    sizeOfy=len(dfY)
    
    stand=(((sizeOfx-1)*(stdX**2))+((sizeOfy-1)*(stdY**2)))/(sizeOfx+sizeOfy-2)

    dOFf=(sizeOfx+sizeOfy-2)

    error=math.sqrt((stand/sizeOfx)+(stand/sizeOfy))

    t_test=((meanX-meanY)-0)/error
    return t_test
```

```python
#point a
dfMales = df[df['Gender'] == 'Male']
dfMales
```

```python
#point a
dfFemales = df[df['Gender'] == 'Female']
dfFemales
```

```python
#test using my function and online calculator 
#full dataframe
Ttest_df(dfMales,dfFemales)
#The p-value is .20916.
#The result is not significant at p < .05.
```

```python
#test using scipy to check
t_val,p_val=ttest_ind(dfMales.Salary,dfFemales.Salary);t_val,p_val #full
```

my null hypo is H0= Mean(Male)- Mean(Female)=0
my alter hypo is H1= Mean(Male)- Mean(Female)!=0
I will set the varince by : unknown but assum to be equal,
and this two new sub-datasets are independant .
I will use T test because of the type of the variance and the sub-datasets

```python
#point b
dfMales=dfMales.sort_values('Age')
dfMalesUnder = dfMales[dfMales['Age'] < 35]
dfMalesOver = dfMales[dfMales['Age'] >= 35]
dfMalesOver
```

```python
#test using my function and online calculator 
#under dataframe
Ttest_df(dfMalesUnder,dfFemalesUnder)
#The p-value is .816559.
#The result is not significant at p < .05.
```

```python
#test using scipy to check
t_valU,p_valU=ttest_ind(dfMalesUnder.Salary,dfFemalesUnder.Salary);t_valU,p_valU #under
```

```python
#point b
dfFemales=dfFemales.sort_values('Age')
dfFemalesUnder = dfFemales[dfFemales['Age'] < 35]
dfFemalesOver = dfFemales[dfFemales['Age'] >= 35]
dfFemalesOver
```

```python
#test using my function and online calculator 
#over dataframe
Ttest_df(dfMalesOver,dfFemalesOver)
#The p-value is .15937.
#The result is not significant at p < .05.
```
